```python
# O que vamos utilizar? 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Qual serão os dados? Quem é x e quem é y no nosso modelo? Quantas amostras temos no nosso modelo?
dataset = pd.read_csv('Salary_Data.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
#quem vai para o treino e quem vai para o teste?
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state=0)
plt.scatter(x_train,y_train)
plt.title("Experiência vs. Salário")
plt.xlabel("experiência (anos)")
plt.ylabel("salário anual")
```

```python
# Vamos criar o modelo de regressão? E agora, vamos treiná-lo?
model = LinearRegression()
model.fit(x_train, y_train)
#Vamos visualizar o modelo?
plt.scatter(x_train,y_train)
plt.plot(x_train,  model.predict(x_train), color='magenta')
plt.title("Experiência vs. Salário (Treino)")
plt.xlabel("experiência (anos)")
plt.ylabel("salário anual")
```

```python
#Agora que temos o modelo treinado, vamos testá-lo?
y_estimated = model.predict(x_test)
```
